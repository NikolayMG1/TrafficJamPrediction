```python
df = df[df['unit'] == 'km/h']

df = df.reset_index(drop=True)

print(df.shape)
```

Check if Speed is a valid number.

```python
df['value'] = pd.to_numeric(df['value'], errors='coerce')

df = df[(df['value'].notna()) & (df['value'] >= 0) & (df['value'] < 150)]

df = df.reset_index(drop=True)

print(df.shape)
```
